```python
km = 8.3
print("kilometers:",km)

con_f = 0.62137

miles = km * con_f
print("\nAfter converting kilometers to miles: ",miles)
```

```python
# 2. Write a Python program to convert Celsius to Fahrenheit?
```

```python
celsius = 35.5
print("celsius:",celsius)

fahrenheit = (celsius * 1.8) + 32

print("\nAfter coverting celsius to fahrenheit:",fahrenheit)
```
